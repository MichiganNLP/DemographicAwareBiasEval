```python
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')  
import torch.nn as nn 
from transformers import BertForSequenceClassification, AdamW, BertConfig, BertForMaskedLM, BertModel

# Load BertForSequenceClassification, the pretrained BERT model with a single 
# linear classification layer on top. 
model = BertModel.from_pretrained(
    "bert-base-uncased", 
    num_labels = 20, 
    output_attentions = False, 
    output_hidden_states = True, 
)

# checkpoint = torch.load("bert_unmasked_trainnew_3.pth")

model = BertModel.from_pretrained("Models/Model1/")#, local_files_only=True)
tokenizer = BertTokenizer.from_pretrained('Models/Model1/')  

file = open('Models/Model1/vocab.txt','r')
lines = file.read().split('\n')

embedding_matrix = model.embeddings.word_embeddings.weight
```

```python
import plotly.graph_objects as go  
import math 
from sklearn.decomposition import PCA  
import matplotlib.pyplot as plt  
import numpy as np  
import plotly.graph_objects as go  
from PIL import Image  

f = ['worda']
m = ['wordb']

she_ind = []
he_ind = []


for i in f:
    she_ind.append(lines.index(i))
    
for i in m:
    he_ind.append(lines.index(i))
    
emb_she = []
emb_he = []
for i in she_ind:
    emb_she.append(embedding_matrix[i])
for i in he_ind:
    emb_he.append(embedding_matrix[i])
emmb = []
for v in emb_he:
    emmb.append(v.detach().numpy())

emmb2 = []
for v in emb_she:
    emmb2.append(v.detach().numpy())

  
she_vector = embedding_matrix[2016]
he_vector = embedding_matrix[2002]

# Coordinates of 'he' and 'she' on the axis  
emb_heshe = [he_vector.detach().numpy(), she_vector.detach().numpy()]

pca = PCA(n_components=2)  
emh_heshe_ = pca.fit_transform(emb_heshe)
he_coords, she_coords = emh_heshe_[0,:], emh_heshe_[1,:]  
# he_coords = [0, 0]  
# she_coords = [1, 0]  


embeddings = emmb  + emmb2
words = list(m) + list(f)

pca = PCA(n_components=2)  
embeddings_2d = pca.fit_transform(np.array(embeddings))  

x, y = embeddings_2d[:,0], embeddings_2d[:,1]  
  
# Data for scatter points  
words1 = words
x1 = x  
y1 = y
  
    
distances = []  
for xi, yi in zip(x1, y1):  
    distance_to_he = math.sqrt((xi - he_coords[0])**2 + (yi - he_coords[1])**2)  
    distance_to_she = math.sqrt((xi - she_coords[0])**2 + (yi - she_coords[1])**2)  
    distances.append(distance_to_he-distance_to_she)  
  
# Create the scatter trace  
scatter_trace = go.Scatter(  
    x=x1,  
    y=y1,  
    mode='markers',  
    marker=dict(size=1, color=distances, colorscale='RdBu_r', cmin=min(distances), cmax=max(distances)),  
    text=words1,  
    textposition='middle center',
#     hoverinfo = 'none',
    hovertemplate='(%{x:.2f}, %{y:.2f})<br>%{text}<br>Distance: %{marker.color:.2f}' ,
    showlegend=False
)  


# Create a trace for the labels  
label_trace = go.Scatter(  
    x=x1,  
    y=y1,  
    mode='text',  
    text=words1,  
    textposition='top center',  
    hoverinfo='none',
    textfont=dict(color='black'),
    showlegend=False
)  
  

# fig = go.Figure()  

x_range = [min(x1) - 0.1, max(x1) + 2]  
y_range = [min(y1) - 0.1, max(y1) + 0.1]  


axis_data = ['he', 'she']
# Create the figure layout  
fig_layout = go.Layout( 
    
    xaxis=dict(  
        range= x_range,  
        tickmode='array',  
        tickvals=[he_coords[0], she_coords[0]],  
        ticktext=axis_data,
        linecolor='black',  # Add color to the axis line  
#         tickfont=dict(color='red'),
        
    ),  
    yaxis=dict(  
        range=y_range,  
#         title=  
    ),  
    title='Africa',  
    hovermode='closest'  
)  


he_she_trace = go.Scatter(    
    x=[he_coords[0], she_coords[0]],    
    y=[he_coords[1], she_coords[1]],    
    mode='markers',    
    marker=dict(size=10, color='red'),  # Larger marker size and different color  
    text=axis_data,  
    showlegend=False
)  
    
    
    
# Create the figure and add the scatter trace  
fig = go.Figure(data=[scatter_trace, he_she_trace, label_trace], layout=fig_layout)  

# Read the image file for the custom icon 
icon1 = Image.open('1.png')  
icon2 = Image.open('2.png')  


fig.add_layout_image(  
        source=icon1,  
        x=x[words1.index('worda')],  
        y=y[words1.index('worda')]-0.07,  
        xanchor='center',  
        yanchor='middle',  
        xref='x',  
        yref='y',  
        sizex=0.15,  
        sizey=0.15,  
#         sizing='stretch',  
        opacity=1
    )  
  
fig.add_layout_image(  
        source=icon2,  
        x=x[words1.index('wordb')],  
        y=y[words1.index('wordb')]+0.07,  
        xanchor='center',  
        yanchor='middle',  
        xref='x',  
        yref='y',  
        sizex=0.15,  
        sizey=0.15,  
#         sizing='stretch',  
        opacity=1
    )  
  

# Show the figure  
fig.show()  

import plotly.io as pio  
  
```
